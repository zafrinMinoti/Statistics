```python
import numpy as np
from scipy.stats import norm, uniform
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
fig = plt.figure(figsize=(8,4),dpi=96)
ax = fig.add_axes([0,0,1,1])

l = 0
s = [1,10,100,1000]

for i in s:
    x = np.linspace(norm.ppf(.01, loc=l, scale=i),
                norm.ppf(.99, loc=l, scale=i), 100)
    ax.plot(x, norm.pdf(x, loc=l, scale=i), label='normal {}'.format(i))
    
    #if i == 1000:
    #    ax.plot(x, uniform.pdf(x, loc=l, scale=i), label='uniform pdf')
        
ax.legend(loc='best')
```

```python
fig = plt.figure(figsize=(8,4),dpi=96)
ax = fig.add_axes([0,0,1,1])

l = 0
s = [1,10,100,1000]

x = np.linspace(-10, 10, 1000)

for i in s:
    ax.plot(x, norm.pdf(x, loc=l, scale=i), label='normal {}'.format(i))

ax.plot(x, uniform.pdf(x, loc=0, scale=1000), label='uniform pdf')
        
ax.legend(loc='best')
```
